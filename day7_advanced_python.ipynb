groupby

```python
import pandas as pd
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
        'Person' :['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
                'Sales':[200,120,340,124,243,350]}

df = pd.DataFrame(data)
df
df
```

```python
df.groupby("Company")
```

```python
by_comp = df.groupby("Company")
by_comp
```

```python
by_comp.mean(numeric_only=True)
```

```python
by_comp.max(numeric_only=True)
```

```python
by_comp.std(numeric_only=True)
```

```python
by_comp.count()
```

```python
by_comp.describe()
```

```python
by_comp.describe().transpose()
```

```python
import pandas as pd
```

```python
df1 = {'A': ['A1','A2','A3','A4'],
        'B': ['B1','B2','B3','B4'],
        'C': ['C 1','C2','C3','C4'],
        'D': ['D1','D2','D3','D4']}
df1 = pd.DataFrame(df1, index = [0,1,2,3])
df1
```

```python
df2 = {'A': ['A5','A6','A7','A8'],
        'B': ['B5','B6','B7','B8'],
        'C': ['C 5','C6','C7','C8'],
        'D': ['D5','D6','D7','D8']}
df2 = pd.DataFrame(df2, index = [5,6,7,8])
df2
```

```python
df3 = {'A': ['A9','A10','A11','A12'],
        'B': ['B9','B10','B11','B12'],
        'C': ['C9','C10','C11','C12'],
        'D': ['D9','D10','D11','D12']}
df3 = pd.DataFrame(df3, index = [9,10,11,12])
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3], axis = 1)
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
display(right)
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

display(left)
```

```python
left
```

merge

```python
pd.merge(left,right,how='inner',on='key')
```
